```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df.head()
```

```python
df.shape
```

## Univariate Analysis

```python
df.isnull().sum()
```

```python
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(), annot=True)
```

## Univariante Analysis

```python
df_setosa = df.loc[df['species']=='setosa']
df_virginica = df.loc[df['species']=='virginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.ones_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.ones_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
df['sepal_length'].plot.hist(bins = 15, color = 'skyblue')
```

## Bivariant Analysis

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,"sepal_length","sepal_width").add_legend()
```

```python
### Bivariant Analysis
sns.countplot(x = 'sepal_length',hue = 'species', data = df, palette = 'Set2')
```

## Multivariant Analysis

```python
sns.pairplot(df,hue='species',size=5)
```

```python
sns.relplot(x = 'sepal_length', y ='sepal_width', col = 'species', data = df, color = 'green')
```
