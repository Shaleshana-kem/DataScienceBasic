## Data Cleaning 
fill missing data instead of just dropping the missing age data rows.
* filling it with mean age of all passengers(imputation)
* find average age with passenger class

```python
sns.boxplot(x='Pclass',y='Age',data=train)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train["Age"] = train[["Age","Pclass"]].apply(impute_age,axis=1) ## fill with average passenger class age
```

drop the Cabin column 

```python
train.drop('Cabin', axis=1,inplace=True)
```

```python
train
```

```python
pd.value_counts(train["Embarked"])
```

```python
train.dropna(inplace=True)
```

```python
train.count()
```

### Converting Categorical Features

Convert Categorical Features to dummy variable using pandas

```python
pd.get_dummies(train["Embarked"],drop_first= True).head()
```

```python
embarked =pd.get_dummies(train["Embarked"],drop_first= True)
sex=pd.get_dummies(train["Sex"],drop_first= True)
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train.head()
```

```python
train = pd.concat([train,sex,embarked],axis=1)
train.head()
```

## Building Logistics Regression Model
starting with splitting our data into training set and test set 

## Sklearn Library
Simple and efficient tools for predictive data analysis

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train.drop(['Survived'],axis=1),train['Survived'],test_size=0.25,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
acc= confusion_matrix(Y_test,pred)
```

```python
acc
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,pred)
```
