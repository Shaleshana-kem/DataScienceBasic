## Matplotlib library

visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. 

```python
import matplotlib.pyplot as plt

%matplotlib inline #eliminate the need to use plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# plt scatter

plt.scatter(x,y,c='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph in 2D")
plt.savefig("Test.png")
```

```python
#plt plot
plt.plot(x,x**3,'r--')
plt.plot(x,x**3,'ro')
```

```python
# plt subplot
plt.subplot(2,2,2)
plt.plot(x,x**3)
plt.subplot(2,2,1)
plt.plot(x,x**2)
plt.subplot(2,2,3)
y=np.arange(0,10,0.1)
plt.plot(y,np.sin(y))
```

```python
# bar plot

x =[2,8,10]
y= [11,16,9]

x1 =[4,8,10]
y1= [11,15,9]
plt.bar(x,y)
plt.bar(x1,y1,color='r')

```

```python
# histograms

a= np.array([22,87,5,10,55,60,45,78,34,8,0,12,23,28])
plt.hist(a
        )
```

```python
#box plot

data = [np.random.normal(0,std,100) for std in range(5,10)]

plt.boxplot(data)  # percentile
plt.show()
```

```python
# pie chart

label= 'python', "c++", "Ruby", "R"
sizes = [214,150,170,200]
colors =  ["green","yellow","blue","red"]
explode =(0.1,0,0,0)
plt.pie(sizes,explode=explode,colors=colors,labels=label, autopct="%1.1f%%")
```

# Seaborn Tutorial

Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Distribution plots
* distplot
* joinplot
* pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

# correlation with heatmap

A correlation heatmap uses colored cells(monochroomatic scale to show 2D correlation matrix),its important in Feature Selection

```python
sns.heatmap(df.corr())
```

# JointPlot

allows to study the relationship between 2 numeric variables.(Univariant Analysis)

```python
sns.jointplot(x = 'tip', y = 'total_bill', data=df, kind='hexa')
```

# PairPlot
known as scatter plot, shows all variable paired with all the other

```python
sns.pairplot(df)
```

Flexibly plot a univariate distribution of observations.

```python
sns.distplot(df["tip"])
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Catergorical Data points.
* boxplot
* violinplot
* countplot
* bar plot

```python
## Count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.barplot(x='smoker',y='tip',data=df)
```

### Box plot

A box and whisker plot is graph that representation from a five number summary

```python
sns.boxplot(x='smoker',y='tip',data=df)
```

```python
sns.boxplot(x='smoker',y='tip',hue="day",data=df)
```

### Violin plot

Violin plot helps us to seeboth the distribution of data in terms of kernel density estimation and box plot

```python
sns.violinplot(x='smoker',y='tip',hue="day",data=df)
```

```python
plt.violinplot(df['size'])
```
